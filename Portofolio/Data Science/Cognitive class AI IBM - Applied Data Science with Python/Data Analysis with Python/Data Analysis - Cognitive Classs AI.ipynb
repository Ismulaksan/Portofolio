## Modul 2 - Data Wrangling

```python
# konversi "mpg" to L/100km pada dataset

df["city-mpg"] = 235/df["city-mpg"]

df.rename(columns={"city-mpg": "city-L/100km"}, inplace=True)
```

```python
#menghilangkan missing value kolom "price"

df.dropna(subset=["price"], axis=0)
```

```python
# Mengganti missing value

import numpy as np

df.replace("?", np.nan, inplace=True)
df.head(5)
```

```python
# Evaluasi Missing Data

missing_data = df.isnull()
missing_data.head()
```

```python
# Menghitung missing data tiap kolom

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

Terdapat 7 variabel/atribute yang memiliki missing data

1. "normalized-losses" : 41 missing data
2. "num-of-doors" : 2 missing data
3. "bore" : 4 missing data
4. "stroke" : 4 missing data
5. "horsepower" : 2 missing data
6. "peak-rpm" : 2 missing data
7. "price": 4 missing data

variabel "num-of-door" : 2 missing data diganti dengan empat karena mayoritas data 84% mobil sedan memiliki 4 pintu
variabel "price" : 4 missing data, di delete semua row karena variabel price merupakan data prediksi.
sedangkan variabel lainnya missing value diganti dengan nilai mean atau rata-rata

```python
# Menghitung rata-rata dari variabel

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print ("Average of normalized:", avg_norm_loss)
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "normalized-losses"

df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Menghitung mean variabel

avg_bore = df["bore"].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "bore"

df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Menghitung mean variabel

avg_stroke = df["stroke"].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "stroke"

df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
# Menghitung mean variabel

avg_horsepower = df["horsepower"].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "horsepower"

df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# Menghitung mean variabel

avg_peakrpm = df["peak-rpm"].astype('float').mean(axis=0)
print("Average of peakrpm:", avg_peakrpm)
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "horsepower"

df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# Menghitung frekuensi variabel

df["num-of-doors"].value_counts().idxmax()
```

```python
# Mengganti missing value dengan nilai mean/rata-rata pada variabel "horsepower"

df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# Menghilangkan features atau row Nan pada variabel price

df.dropna(subset=["price"], axis=0, inplace=True)

# Mereset indeks
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
# Melihat tipe data tiap variabel

df.dtypes
```

```python
# Mengubah format variabel
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

<h2>Melakukan normalisasi data</h2>

```python
df.rename(columns={"highway-mpg": "highway-L/100km"}, inplace=True)
```

```python
# Mengganti (original value) by (original value)/(maximum value)

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 


# show the scaled columns
df[["length","width","height"]].head()
```

## Binning data 

```python
# Mengganti tipe data variabel horsepower

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# Membuat grafik distribusi data variabel horsepower

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Membuat array dengan kategori nilai maksimal dan nilai minimal serta membagi 3 kategori (4 pembagi)

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# Membuat kategori dari variabel horsepower

group_names = ['Low', 'Medium', 'High']
```

```python
# Membuat variabel horsepower-binned

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
# Menghitung jumlah kategori

df["horsepower-binned"].value_counts()
```

```python
#Grafik kategori pada horsepower-binned

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
