```python
# Melakukan grouping pada variabel drive-wheels, bdy style dan price

df_group_1 = df[["drive-wheels", "body-style", "price"]]
df_group_1 = df_group_1.groupby(["drive-wheels"], as_index=False).mean()
df_group_1
```

```python
df_gptest = df[["drive-wheels", "body-style", "price"]]
group_test1 = df_gptest.groupby(["drive-wheels", "body-style"], as_index=False).mean()
group_test1
```

```python
# Menggunakan pivot tabel pada variabel drive-wheels dan body-style

group_pivot = group_test1.pivot(index="drive-wheels", columns="body-style")
group_pivot
```

```python
group_pivot = group_pivot.fillna(0)
group_pivot
```

```python
# Melihat korelasi antar variabel drive-wheels body style dan price dengan menggunakan heatmap

plt.pcolor(group_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
# Memberikan keterangan dan label pada heatmap

import numpy as np

fig, ax = plt.subplots()
im = ax.pcolor(group_pivot, cmap = "RdBu")

#Label names
row_labels = group_pivot.columns.levels[1]
col_labels = group_pivot.index

# Memindahkan label dan ticks ke tengah
ax.set_xticks(np.arange(group_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(group_pivot.shape[0]) + 0.5, minor=False)

#Insert Label
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
# Melakukan uji ANOVA pada variabel

f_val, p_val = stats.f_oneway(group_test2.get_group('fwd')['price'], group_test2.get_group('rwd')['price'], group_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(group_test2.get_group('fwd')['price'], group_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```
